```python
import os
import glob
import pathlib
import cv2
import numpy as np
from PIL import Image
import pdb


cwd= os.getcwd()
print('getcwd:      ', os.getcwd())
#  /home/takahashi/takahashi/11class
print("cwd  ",cwd)#pjynbファイルのディレクトリを取得

# files = glob.glob(cwd)#mp4拡張子のファイル名を全て取得
files = os.listdir(cwd)#mp4拡張子のファイル名を全て取得

print("files = ",files)

#画像のあるフォルダリスト格納用リスト
pic_lists = []
#/home/takahashi/takahashi/11class/Asahigaoka/base_movie
#拡張子付きで出力されるので注意
for file in files:
    print("file= ",file)
#     print('check file dir =',cwd + "\\" + file )
#     print(cwd + "\\" + file + "\\" + '*.jpg')
#     if os.path.exists( cwd + "\\" + file + "\\" + '*.jpg'):
#     print("file  ",file)
    # lists = glob.glob(file + '/*.jpg')
    lists = glob.glob(file + '/*.png')
    print("lists =  ",lists)
    if len(lists) > 0:
        pic_lists.append(file)
        get_fname = pathlib.Path(file)
#         print("get_fname.stem  ",get_fname.stem)#拡張子を除いたファイル名を出力
print("pic_lists =",pic_lists)

# def find_png_files():
#     # 現在のディレクトリを取得
#     current_directory = os.getcwd()

#     # 現在のディレクトリ以下のすべての.pngファイルを検索
#     png_files = glob.glob(current_directory + '/**/*.png', recursive=True)

#     return png_files

# # 関数を実行して結果を表示
# pic_lists = find_png_files()
# print(pic_lists)
```

```python
def pic_gen():
    global cwd,get_fname
    DATADIR =cwd# "./Asahigaoka_study_arinashi"
    # CATEGORIES = ["0_80", "80_"]
    CATEGORIES ="/"+str(get_fname)#'/80_'  #"0_80/*.jpg"
    path=cwd+CATEGORIES
 # 入力画像の保存先パス
    input_path = path
    # files = glob.glob(input_path + '/*.jpg')
    pics = glob.glob(input_path + '/*.jpg')
    
    # 出力画像の保存先パス
    output_path = str(path)+ "_gen"#"/output/"
    
    if os.path.isdir(output_path) == False:
        os.mkdir(output_path)
        
    print("generating"+str(get_fname))
    
    for i, pic in enumerate(pics):
        i=i+1
        # 画像の読み込み
#         img = cv2.imread(pic)
        shifted_img = np.array(Image.open(pic))
        #画像サイズ取得
        h, w, c = shifted_img.shape
        print(h,w,c)
#         img = load_img(pic)
        # どれだけ画像をシフトさせるか
        x_diff = 0
        y_diff = -200

        # 画像をシフト
#         shifted_img = np.roll(img, shift=(y_diff, x_diff), axis=(0, 1))
        # はみ出た部分を削除 (黒塗り)

        if x_diff >= 0:
            shifted_img[:, :x_diff] = 0
        else:
            shifted_img[:, w + x_diff:] = 0

        if y_diff >= 0:
            shifted_img[:y_diff] = 0
        else:
            shifted_img[h + y_diff:] = 0
        
        #画像を保存
        shifted_img = cv2.cvtColor(shifted_img, cv2.COLOR_RGBA2BGRA)
        cv2.imwrite(output_path+'/shifted_'+str(i)+'.jpg',shifted_img)
```

```python
#Main Loop
files = os.listdir(cwd)#すべてののファイル名を全て取得
print(files)
print(pic_lists)
#拡張子付きで出力されるので注意
for pic_list in pic_lists:
    if os.path.isdir(pic_list): 
#     if file.is_dir():
        print("pic_list  ",pic_list)
        get_fname = pathlib.Path(pic_list)
        print("get_fname.stem  ",get_fname.stem)#拡張子を除いたファイル名を出力
        pic_gen()
    
print("generate_finish!")
```
